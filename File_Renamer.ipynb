```python
# Rename for Numbers


def rename_file(folder_path, old_filename_prefix, folder_name):
    """
    Renames a file starting with a specific prefix within a folder.

    Args:
        folder_path (str): The path to the folder containing the file to rename.
        old_filename_prefix (str): The starting prefix of the file to rename (e.g., "20240704").
        new_filename_prefix (str): The desired starting prefix for the renamed file (e.g., "W12P54").
    """

    for filename in os.listdir(folder_path):
        if filename.startswith(old_filename_prefix):
            # Assuming consistent formatting after the prefix
            new_filename = f"N{int(folder_name)+1}P68.xef" # Enter the person number 
            old_filepath = os.path.join(folder_path, filename)
            new_filepath = os.path.join(folder_path, new_filename)
            os.rename(old_filepath, new_filepath)
            print(f"Renamed '{filename}' to '{new_filename}' in folder '{folder_path}'.")
            return  # Exit the loop after one rename

    print(f"No file starting with '{old_filename_prefix}' found in folder '{folder_path}'.")

if __name__ == "__main__":
    parent_folder_path = r"E:\ISL_DATASET\Numbers"  # Replace with your actual path
    old_filename_prefix = "20240805" # Rename with the prefix you need

    # Iterate through each subfolder (assuming folder names)
    for folder_name in os.listdir(parent_folder_path):
        folder_path = os.path.join(parent_folder_path, folder_name)
        if os.path.isdir(folder_path):  # Check if it's a directory
            rename_file(folder_path, old_filename_prefix, folder_name)
```
